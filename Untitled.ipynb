```python
size = [2441, 2312, 1031, 1233]
plt.rc('font', family='Malgun Gothic')
plt.axis('equal') # axis축을  더욱더 둥글게 만듬
label = ['A형','B형','AB형','O형'] #라벨을 지정후
plt.pie(size, labels=label, autopct='%.2f%%') #라벨을 적용,  autopct  자동으로 퍼센테이지 지정 %.n%% n 번쨰 까지 지정
plt.show()
```

```python
size = [2441, 2312, 1031, 1233]
plt.rc('font', family='Malgun Gothic')
plt.axis('equal') # axis축을  더욱더 둥글게 만듬
label = ['A형','B형','AB형','O형'] #라벨을 지정후
color = ['skyblue', 'deeppink', 'hotpink', 'pink'] # 색상 
 #라벨을 적용,  autopct  자동으로 퍼센테이지 지정 %.n%% n 번쨰 까지 지정, eexplode 지정해줌
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode = (0.3,0,0,0))
plt.show()
```

```python
import csv
f= open('gender.csv')
data = csv.reader(f)
size = []

name = input('찾고 싶은 지역')
for row in data:
  if name in row[0]:
    m = 0
    f = 0
    for i in range(101):
        m += int(row[i+3].replace(',',''))
        f += int(row[i+106].replace(',',''))
    break
size.append(m)
size.append(f)
print(size)
a = 0
b = 0 
if size[0] > size[1]:
    a = 0.1
else:
    b = 0.1
plt.rc('font', family='Malgun Gothic')
plt.axis('equal') # axis축을  더욱더 둥글게 만듬
color = ['skyblue', 'hotpink']
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color,explode=(a,b), startangle=90) # startangle -> 시작 각도를 지정해줌
plt.title(name+ '지역의 남녀 성별 비율')
plt.show()
```
