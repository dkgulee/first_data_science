```python
size = [2441, 2312, 1031, 1233]
plt.rc('font', family='Malgun Gothic')
plt.axis('equal') # axis축을  더욱더 둥글게 만듬
label = ['A형','B형','AB형','O형'] #라벨을 지정후
color = ['skyblue', 'deeppink', 'hotpink', 'pink'] # 색상 
 #라벨을 적용,  autopct  자동으로 퍼센테이지 지정 %.n%% n 번쨰 까지 지정, eexplode 지정해줌
plt.pie(size, labels=label, autopct='%.1f%%', colors=color, explode = (0.3,0,0,0))
plt.show()
```

```python
import csv
f= open('gender.csv')
data = csv.reader(f)
size = []

name = input('찾고 싶은 지역')
for row in data:
  if name in row[0]:
    m = 0
    f = 0
    for i in range(101):
        m += int(row[i+3].replace(',',''))
        f += int(row[i+106].replace(',',''))
    break
size.append(m)
size.append(f)
print(size)
a = 0
b = 0 
if size[0] > size[1]:
    a = 0.1
else:
    b = 0.1
plt.rc('font', family='Malgun Gothic')
plt.axis('equal') # axis축을  더욱더 둥글게 만듬
color = ['skyblue', 'hotpink']
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', colors=color,explode=(a,b), startangle=90) # startangle -> 시작 각도를 지정해줌
plt.title(name+ '지역의 남녀 성별 비율')
plt.show()
```

```python
#꺾인선 그래프로 표현하기
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

name = input('궁금한 동네를 입력해주세요')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break
plt.plot(m, label='Male')
plt.plot(f, label='Female')
plt.legend()
plt.show()
```

```python
# 남자 인구와 여자인구의 비교
f = open('gender.csv')
data = csv.reader(f)
result = []
name = input('궁금한 동네를 입력해주세요 :')
for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i].replace(',', '')) - int(row[i+103].replace(',', '')))
        break
plt.bar(range(101), result)
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.show()
```

```python
# matplotlib 에서 사용할 수 있는 다양한 컬러 를 cmap 에 입력
x =[]
y =[]
size = []
for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
plt.scatter(x, y, s=size, c = size, cmap='flag')
plt.colorbar()
plt.show()
```

```python
# matplotlib 에서 사용할 수 있는 다양한 컬러 를 cmap 에 입력
# alpha 투명도를 할당 해줌 0에 가까울 수록 투명색에 가까워짐
x =[]
y =[]
size = []
for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(10,100))
plt.scatter(x, y, s=size, c = size, cmap='flag', alpha = 0.1)
plt.colorbar()
plt.show()
```

```python
# 제주도의 연령대별 성별 비율을 산점도로 표현하기
f = open('gender.csv')
data = csv.reader(f)
m = []
f = []
name = input('궁금한 동네를 입력해주세요 :')
for row in data:
    if name in row[0]:
        for i in range(3, 104):
            m.append(int(row[i].replace(',','')))
            f.append(int(row[i+103].replace(',','')))
        break
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize= (10,5) , dpi= 300)
plt.title(name +'지역의 성별 인구 그래프')
plt.scatter(m ,f, s= size, c= range(101), alpha=0.5, cmap='jet')
plt.colorbar()
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```
