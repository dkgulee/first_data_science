```python
import csv
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
max_temp = -999
max_date = ''
for row in data:
    if row[-1] == '' :
        row[-1] = -999
    # 최고기온을 실수형태로 변경
    row[-1] = float(row[-1])
    if max_temp < row[-1]:
        max_date = row[0]
        max_temp = row[-1]
f.close()
print(max_date, max_temp)
print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은',max_date+'로',max_temp,'도 였습니다.')
```

```python
# 1983 년의 이후 가장 높은 기온과 가장 낮은 기온 2월 14일의 최고 최저 기온을 저장
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
high = []
low = []

for row in data:
    if row[-1] != '' and row[-2] != '':
         if int(row[0].split('-')[0]) >= 1983:
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.plot(high, 'hotpink' ,label='high')
plt.plot(low, 'skyblue', label='low')
plt.title('내 생일의 기온 변화 그래프')
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
plt.legend()
plt.show()
```

```python
#내생일의 가장높은 기온과 가장낮은 기온을 찾아라
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)
max = []
min = []


for row in data:
    day = row[0].split('-')
    if row[-1] != '' and row[-2] != '':
        if int(day[0]) >= 1993:
            if day[1] == '12' and day[2]=='21':
                max.append(float(row[-1]))
                min.append(float(row[-2]))
plt.plot(max, 'red' , label = 'max')
plt.plot(min, 'skyblue', label = 'min')
plt.title('내생일별 온도')
plt.legend()
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='Malgun Gothic')
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)


# 8월 일별 기온 데이터를 상자그림으로 표현하기

hday =[i*[] for i in range(31)]
lday =[i*[] for i in range(31)]

for row in data:
    if row[-1] != '' and row[0].split('-')[1] == '08':
        hday[int(row[0].split('-')[2])-1].append(float(row[-1]))
        lday[int(row[0].split('-')[2])-1].append(float(row[-1]))
plt.style.use('ggplot')
plt.figure(figsize=(10, 5), dpi= 300)
plt.boxplot(lday)
plt.boxplot(hday, showfliers = False) # showfliers => 이상치를 제거해서 보여줌
plt.show()
```

```python
f = open('seoul.csv')
data = csv.reader(f)
header = next(data)

#1년 중 일교차가 가장 큰 달은  몇월인가요?
max_daily_range =  -99999
max_row =  0
for row in data:
    if row[-1] != '' and row[-2] != '':
        temp = abs(float(row[-1])- float(row[-2]))
        if max_daily_range < temp:
            max_daily_range = temp
            max_row = row

print(max_daily_range)
print(max_row[0].split('-'))
        
```
