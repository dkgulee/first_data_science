```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0., 5., 0.2)
print(t)
plt.plot(t, t,'r--',t**2, 'bs',t,t**3,'g^')
plt.show()
```

```python
import matplotlib.pyplot as plt
t = []
p2 =[]
p3 =[]
for i in range(0, 50, 2):
    t.append(i/10)
    p2.append((i/10)**2)
    p3.append((i/10)**3)
plt.plot(t, t, 'r--', t, p2, 'bs', t, p3, 'g^')
plt.show()
```

```python
# 2의 제곱근을 출력
print(numpy.sqrt(2))
print(np.sqrt(2))
# 파이를 출력
print(np.pi)
# sin(0)을 출력
print(np.sin(0))
# cos(pi) 출력
print(np.cos(np.pi))
# numpy 에 내장되어 있는 random을 호출
print(np.random.rand(5))
type(np.random.rand(5))
print(np.random.choice(6,10)) # 0에서 5사이의 랜덤값을 10개 호출 복원추출
print(np.random.choice(10,6, replace =False)) # 0에서 9사이의 랜덤값 비복원 추출
#확률을 지정해서 나타낼수 있다
print(np.random.choice(6,10, p=[0.1,0.2,0.3,0.2,0.1,0.1]))
```

```python
# 그냥 버블차트 그리기
x = []
y = []
size = []

for i in range(200):
    x.append(random.randint(10, 100))
    y.append(random.randint(10, 100))
    size.append(random.randint(10, 100))

plt.scatter(x, y, s=size, c= x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()

# nump 로 버블차트 그리기
x = np.random.randint(10, 100, 200)
y = np.random.randint(10, 100, 200)
size = np.random.rand(100) * 100

plt.scatter(x, y, s=size, c= x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
a = np.array([1,2,3,4]) # numpy는 arrey로 만들면 , 없다
b = [1,2,3,4] # list 는 ,이 있다
print(b)
print(a)
print(a[1:]) # 슬라이싱
print(a[1],a[-1]) #인덱싱으로 접근이 가능함
```

```python
#배열을 생성하기 윟 꼭 리스트가 있어야하는 것은 아닙니다. zeros(), ones(), eye()같은
# 다양한 함수를 사용해서 배열을 초기화 가능
a = np.zeros(10)
print(a)
b = np.ones(10)
print(b)
c = np.eye(4)
print(c)
# 연속된 숫자로 데이터를 생성하는 것
print(np.arange(3))   # 0 ~ n 까지
print(np.arange(3,7))  # n ~ m-1 까지
print(np.arange(3,7,2)) # n ~ m-1 까지 l만큼 간격을 둠

a = [0.9+round(float(i*0.1),2) for i in range(1, 11)]
print(a)
a = np.arange(1, 2,0.1) # 1이상 2미만 구간에서 0.1 간격으로 실수 생성
print(a)
b = np.linspace(1, 2, 11) # 1부터 2까지 11개로 나눈 실수 생성
print(b)

a = np.arange(-np.pi,np.pi,np.pi/10)
b = np.linspace(-np.pi, np.pi, 20)
print(a, b, end='\n\n')
```

```python
a = np.zeros(10) + 5
print(a)
# 값이 5인 배열 100개
b = np.zeros(100) + 5
print(b)
a = np.linspace(1,2, 11)
print(np.sqrt(a))
```

```python
#배열에 어던 연산이나 함수를 적용하면 배열의 모든 값이 한꺼번에 계산이 됨
# 평행이동의 값도 쉽게 구할 수 있다.
a = np.arange(-np.pi, np.pi, np.pi/100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a+np.pi/2, np.sin(a))
plt.show()
```
